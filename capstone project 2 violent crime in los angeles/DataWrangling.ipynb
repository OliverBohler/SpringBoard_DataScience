```python
crimes_2020_2024 = '/Users/oliverbohler/Desktop/Springboard/capstone project 2 violent crime in los angeles/Crime_Data_From_2020_to_Present.csv'

crimes_20_to_present = pd.read_csv(crimes_2020_2024)

crimes_20_to_present.head()
```

```python
crimes_20_to_present.dtypes
```

```python
combined_reports = pd.concat([crimes_10_to_19, crimes_20_to_present], ignore_index= True)
```

```python
combined_reports.head()
```

```python
combined_reports.shape
```

```python
combined_reports.groupby('AREA')['AREA NAME'].value_counts()
```

```python
all_crime = combined_reports.groupby('Crm Cd')['Crm Cd Desc'].value_counts()
```

```python
robbery_df = combined_reports[combined_reports['Crm Cd Desc'] == 'ROBBERY']

# Group by 'Crm Cd' and get value counts for 'Crm Cd Desc'
robbery_counts = robbery_df.groupby('Crm Cd')['Crm Cd Desc'].value_counts()

print(robbery_counts)
```

```python
# Here we want to get an idea of the crimes that are mostly reported and it looks like robberies and burglaries are top

crime_counts_df = all_crime.reset_index( name = 'Counts')
crime_counts_sorted_df = crime_counts_df.sort_values(by='Counts', ascending=False)

print(crime_counts_sorted_df.head(30))
```

```python
# Ok great, we can see that the biggest problem seems to be the amount of thefts, robberies and burglaries 
#let's change some names of columns and drop columns we are not interested in

combined_reports.rename(columns = {'Crm Cd' : 'Crime Code', 'Crm Cd Desc': 'Crime Description'}, inplace = True)
```

```python
# Let's also drop columns that we are not interested in to make it easier

combined_reports = combined_reports.drop(columns =['Part 1-2','Mocodes','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4'])

combined_reports.head()
```

```python
# Sort the dataframe so it only includes crimes that involve the following codes: 

combined_df=combined_reports[combined_reports['Crime Code'].isin([210,310,330,331,420,440,220,320])]
combined_df.head()
```

```python
#Now that we have our dataframe down to the crimes that affect most residents and home owners; let's drop missing values

combined_df.dropna()
combined_df.head()

```

```python
#Check for duplicates and drop them if neccessary 

duplicates = combined_df[combined_df.duplicated()]
if duplicates.empty:
    print("No duplicates found")
else :
    print("Duplicates found")
    print(duplicates)
    
combined_df_unique = combined_df.drop_duplicates()
print(combined_df_unique)
```

```python
#Save DataFrame to csv file

combined_df.to_csv('theft_and_burglary.csv', index = False)
```
