```python
import pandas as pd
import matplotlib.pyplot as plt

crimes_2010_2019 = '/Users/oliverbohler/Desktop/Springboard/capstone project 2 violent crime in los angeles/Crime_Data_From_2010_to_2019.csv'

crimes_10_to_19 = pd.read_csv(crimes_2010_2019)

crimes_10_to_19.head()
```

```python
crimes_10_to_19.info()
```

```python
crimes_2020_2024 = '/Users/oliverbohler/Desktop/Springboard/capstone project 2 violent crime in los angeles/Crime_Data_From_2020_to_Present.csv'

crimes_20_to_present = pd.read_csv(crimes_2020_2024)

crimes_20_to_present.head()
```

```python
crimes_20_to_present.dtypes
```

```python
combined_reports = pd.concat([crimes_10_to_19, crimes_20_to_present], ignore_index= True)
```

```python
combined_reports.head()
```

```python
combined_reports['DATE OCC']= pd.to_datetime(combined_reports['DATE OCC'])
combined_reports['Date Rptd'] = pd.to_datetime(combined_reports['Date Rptd'])
combined_reports['Vict Descent'] = combined_reports['Vict Descent'].astype(str)
combined_reports.dtypes
```

```python
combined_reports.shape
```

```python
combined_reports.groupby('AREA')['AREA NAME'].value_counts()
```

```python
all_crime = combined_reports.groupby('Crm Cd')['Crm Cd Desc'].value_counts()
```

```python
robbery_df = combined_reports[combined_reports['Crm Cd Desc'] == 'ROBBERY']

# Group by 'Crm Cd' and get value counts for 'Crm Cd Desc'
robbery_counts = robbery_df.groupby('Crm Cd')['Crm Cd Desc'].value_counts()

print(robbery_counts)
```

```python
# Here we want to get an idea of the crimes that are mostly reported and it looks like robberies and burglaries are top

crime_counts_df = all_crime.reset_index( name = 'Counts')
crime_counts_sorted_df = crime_counts_df.sort_values(by='Counts', ascending=False)

print(crime_counts_sorted_df.head(30))
```

```python
# Ok great, we can see that the biggest problem seems to be the amount of thefts, robberies and burglaries 
#let's change some names of columns and drop columns we are not interested in

combined_reports.rename(columns = {'Crm Cd' : 'Crime Code', 'Crm Cd Desc': 'Crime Description', 'Premis Cd': 'Premis Code', 'AREA NAME': 'Area Name', 'DATE OCC': 'Date occ'}, inplace = True)
```

```python
# Let's also drop columns that we are not interested in to make it easier

combined_reports = combined_reports.drop(columns =['Part 1-2','Mocodes','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4'])

combined_reports.head()
```

```python
# Sort the dataframe so it only includes crimes that involve the following codes: 

combined_df=combined_reports[combined_reports['Crime Code'].isin([210,310,330,331,420,440,220,320])]
combined_df.head()
```

```python
#Now that we have our dataframe down to the crimes that affect most residents and home owners; let's drop missing values

combined_df.dropna()
combined_df.head()
```

```python
#Check for duplicates and drop them if neccessary 

duplicates = combined_df[combined_df.duplicated()]
if duplicates.empty:
    print("No duplicates found")
else :
    print("Duplicates found")
    print(duplicates)
    
combined_df_unique = combined_df.drop_duplicates()
```

```python
#show the list of missing values
missing_values = pd.concat([combined_df.isnull().sum(), 100 * combined_df.isnull().mean()], axis = 1)
missing_values.columns = ['count','%']
missing_values.sort_values(by = 'count', ascending = False)
                           
```

```python
combined_df = combined_df.drop(columns =['AREA'])
combined_df.head()
```

```python
# Cleaning the Victim Age Column

not_null = (combined_df['Vict Age'] == 0).sum()
print(f"Number of records with age 0: {not_null}")

missing_age = combined_df['Vict Age'].isnull().sum()
print(f"Number of missing values in the age column: {missing_age}")

age_info = (combined_df['Vict Age'] != 0).sum()
print(f"Number of available age info in the column: {age_info}")
```

```python
age_zero = combined_df[combined_df['Vict Age']==0]
age_not_zero = combined_df[combined_df['Vict Age'] != 0]

zero_counts = age_zero['Area Name'].value_counts()
not_zero_counts = age_not_zero['Area Name'].value_counts()

counts = pd.DataFrame({'Age 0': zero_counts, 'Age not 0': not_zero_counts})
plt.figure(figsize=(10,6))
counts.plot(kind ='bar')
plt.title("Values per Area")
plt.xlabel('Area Name')
plt.ylabel('Counts')
plt.show()
```

```python
#The Data looks fairly good, the one Area that has too many 0 entries is Hollenbeck so we will remove it
combined_df = combined_df.loc[~combined_df['Area Name'].isin(['Hollenbeck'])]                                       
```

```python
#we see that there are a few outliers here and we were right to check for the null values in Victim Age.
max_age = combined_df['Vict Age'].max()
print(max_age)
age = combined_df['Vict Age'].value_counts()
print(age)
```

```python
#Only include an age range from 18 to 99 for accuracy
combined_df = combined_df.loc[(combined_df['Vict Age'] > 18) & (combined_df['Vict Age'] <= 99)]
age = combined_df['Vict Age'].value_counts()
print(age)
```

```python
#Check the Vict Descent Column to clean it
descent = combined_df['Vict Descent'].value_counts()
print(descent)
```

```python
# We are only interested in the main 5 being: Hispanic, White, Black, Asian, Other
main5 =['H','W','A','B','O']
combined_df = combined_df[combined_df['Vict Descent'].isin(main5)]
descent = combined_df['Vict Descent'].value_counts()
print(descent)
```

```python
#Let's also inspect the Vict Sex Column
vict_sex = combined_df['Vict Sex'].value_counts()
print(vict_sex)
```

```python
# We filter for M and F data
main_sex =['M','F']
combined_df = combined_df[combined_df['Vict Sex'].isin(main_sex)]
ms = combined_df['Vict Sex'].value_counts()
print(ms)
```

```python
incident_counts = combined_df['Premis Desc'].value_counts().reset_index()

# Rename columns for clarity
incident_counts.columns = ['Premis Desc', 'Incident Count']

# Sort by 'Incident Count' in descending order and get the top 10
top_10_incident_locations = incident_counts.head(10)

print(top_10_incident_locations)
```
